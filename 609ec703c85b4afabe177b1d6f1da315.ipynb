```python
#check the columns
accident1.columns
```

```python
#Add year column
accident1['YEAR'] = accident1['ACCIDENT DATE'].apply(lambda x: int(x[:4]))
#reduce the columns
accident1 = accident1[['ACCIDENT TIME', 'BOROUGH', 'ZIP CODE', 'LATITUDE',
       'LONGITUDE', 'LOCATION','NUMBER OF PERSONS INJURED',
       'NUMBER OF PERSONS KILLED','COLLISION_ID','YEAR']]
#check null on lat and long
accident1.dropna(subset=['LATITUDE', 'LONGITUDE','LOCATION'], inplace = True)
accident_new = gpd.GeoDataFrame(accident1, geometry=gpd.points_from_xy(accident1.LONGITUDE, accident1.LATITUDE))
#change the time to hours
accident_new['ACCIDENT TIME'] = pd.to_datetime(accident_new['ACCIDENT TIME'], format='%H:%M').dt.hour
accident_new = accident_new.drop('LOCATION', axis = 1)
#only consider the injured accidents or killed accident
indexes = accident_new[(accident_new['NUMBER OF PERSONS KILLED'] ==0) & (accident_new['NUMBER OF PERSONS INJURED'] ==0)].index
accident_new = accident_new.drop(index = indexes)
#only consider the year 2019 and 2018, since the dataset is too large
accident_new = accident_new[(accident_new['YEAR']==2019)|(accident_new['YEAR']==2018)]
```

```python
#check the data, and divided the severity
accident_new.groupby('NUMBER OF PERSONS KILLED').count()['ACCIDENT TIME']
```

```python
accident_new.groupby('NUMBER OF PERSONS INJURED').count()['ACCIDENT TIME']
```

```python
#determined the severity of each accident
accident_new = accident_new.assign(severity = 0)
index1 = accident_new[(accident_new['NUMBER OF PERSONS INJURED']==1) & (accident_new['NUMBER OF PERSONS KILLED'] == 0)].index
index2 = accident_new[(accident_new['NUMBER OF PERSONS INJURED']>=2) & (accident_new['NUMBER OF PERSONS INJURED']<=3) & (accident_new['NUMBER OF PERSONS KILLED'] == 0)].index
index3 = accident_new[(accident_new['NUMBER OF PERSONS INJURED']>=4) & (accident_new['NUMBER OF PERSONS INJURED']<=22) & (accident_new['NUMBER OF PERSONS KILLED'] == 0)].index
index4 = accident_new[(accident_new['NUMBER OF PERSONS KILLED'] == 1) |(accident_new['NUMBER OF PERSONS KILLED'] == 2)].index
accident_new.loc[index1,'severity'] = 1
accident_new.loc[index2,'severity'] = 2
accident_new.loc[index3,'severity'] = 3
accident_new.loc[index4,'severity'] = 4
```

```python
#determined the level of time(only selected drinking time according to the background research)
accident_new = accident_new.assign(Drinking_time = 0)
index_night1 = accident_new[(accident_new['ACCIDENT TIME']>=18) & (accident_new['ACCIDENT TIME']<=20)].index
index_night2 = accident_new[(accident_new['ACCIDENT TIME']>=21) & (accident_new['ACCIDENT TIME']<=23)].index
index_night3 = accident_new[(accident_new['ACCIDENT TIME']>=0) & (accident_new['ACCIDENT TIME']<=2)].index
index_night4 = accident_new[(accident_new['ACCIDENT TIME']>=3) & (accident_new['ACCIDENT TIME']<=6)].index
accident_new.loc[index_night1,'Drinking_time'] = 1
accident_new.loc[index_night2,'Drinking_time'] = 2
accident_new.loc[index_night3,'Drinking_time'] = 3
accident_new.loc[index_night4,'Drinking_time'] = 4
```

```python
#final data
accident_new.dropna(subset=['ZIP CODE'], inplace = True)
accident_new.head()
```

## 8. Analysis 
We did four regression models and two random forest regressors to analyse the sale volumes, drinking time vs. severity, total drink vs. number of accidents, number of bars vs. number of accidents, nearest distance of bars vs. accident severity:

- Since we got the data of bars location from the ArcGIS, it contains the sale volume of each bar, we convert the feature layer to the sedf. The sum of sale volume drinking in each zip code area is a huge number, so we decided to calculate the ratio of severity with the total accident in that area  versus total sale volume. After fit into the linear regression model, we got the score for 0.148. It means that the total sale volume of bars is somewhat related to the accident severity. But it shows the more sale volume will have less injuries.
    
- We decided to use aggregate point function to get total drink and accidents for each zip code area separately(using point layer and polygon layer). Because they are in different SEDF, so we merge them into the same SEDF. Then we pick the number of total drinkings and the number of collisions in each zip code area to make a regression model. And we found a positive strong relationship with a score of 0.51. In other words, with the increasing number of drinking, there will be an increasing trend in accidents. 
    
- Because we have a huge number of bars and collisions in each zip code, which makes it harder to see any relationship between them. Also, we believe not all of the car accidents are influenced by alcohol. So we decided to filter out our data, by selecting the point count of bar greater than 10 and accident greater than 200. We believe those data are representative of the relationship between bar and number of accidents. Then, we draw a regression model to analyze the relationship. As we can notice from the model, the number of bars and the number of accidents have a weak relationship with others with the score 0.071. But we can still notice that with more bars in a zip code area, there will be an increasing number of accidents in that region(there are few outliers in this case). 
- We want to consider the time slot with the severity. We make two bar graphs to plot the distribution of time slots and the injury severity.  We could notice the severe injuries increased compared with the time slot. Moreover, we calculate the mean severity for different time regions and we draw a model for this part. From the model we can notice that as you drive later, there will be an increasing level of severity. In the linear regression model, we got a score of 0.74, which means there is a strong relationship between drinking time and accident severity. When people drive later, the injury will be more severe.
- We also want to find whether the nearest distance between the bars and collisions can influence the severity of the accident. We use find_nearest() function to get the distance. However, there is a limitation on the number of dataset, so we select some representative 
- data with the top ten number of bar zip code areas and get 296 rows to use in our model. After this, we make a random forest regressor model to see whether we can use the distance to predict the severity of collision. We also choose some other datas, such as information about the restaurant and market, to predict together with total miles to improve the accuracy rate. We also change the different levels of severity to binary levels and split the dataset to training set and testing set. After running this model, we find out that our model’s R^2 score is -0.43 , which we can know that with further away from the bars, the severity level of collision will be smaller.
